```python
import pandas as pd
import numpy as np
```

```python
url="https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/breast-cancer.csv"
df = pd.read_csv(url)
df.sample(4)
```

```python
df['diagnosis'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df['diagnosis']=labelencoder.fit_transform(df['diagnosis'].values)
df['diagnosis']
```

```python
corrm=df.corr().abs()
corrm.style.background_gradient(cmap='coolwarm')
```

```python
df.corrwith(df['diagnosis']).abs().sort_values(ascending=False)
```

```python
cols=['concave points_worst','perimeter_worst','concave points_mean','radius_worst','perimeter_mean','area_worst','radius_mean','area_mean','concavity_mean','concavity_worst','compactness_mean','compactness_worst']
X=df[cols].copy()
y=df['diagnosis'].copy()
```

```python
from sklearn.preprocessing import StandardScaler
sscaler=StandardScaler()
X=sscaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
y_predict=knn.predict(X_test)
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test,y_predict)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.model_selection import cross_val_predict
prediction=cross_val_predict(estimator=knn,X=X,y=y,cv=9)
print(classification_report(y_test,y_predict))
```
